```python
from dotenv import load_dotenv
load_dotenv()
from MagnusGPTConnector.config import Config
config = Config()
from MagnusGPTConnector.gpt_client import GPTClient
client = GPTClient(config)
```

<h3><b>Calling the Whisper model</b></h3>
Note: Real-time translation is not available

Parameters that can be passed :
<ul>
<li>Audio Path : You can pass the path of the AUDIO file (.wav or .mp3 supported)  </li>
<li>Audio File : You can pass the buffer of the AUDIO file </li>
</ul>

```python
# Whisper - By providing path to audio file (.wav or .mp3)

response = client.invoke_whisper(audio_path='<PATH_TO_AUDIO_FILE>.wav')
print(response)
```

<h3><b>Calling the Embeddings model</b></h3>

<h3><b>Calling the GPT 4o Model</b></h3>

Parameters that can be passed :
<ul>
<li> Input Data : Can receive either file buffer (Only image[.jpg/.jpeg/.png] or pdf file supported) or query/context as a string for GPT </li>
<li> Prompt : Pass any additional instruction for specific task or how you want the context to behave. For eg :output structure needs to be in JSON format</li>
<li> Path To File : Pass the file path (Only image[.jpg/.jpeg/.png] or pdf file supported) directly </li>
</ul>


```python
# GPT-4o, Text based Query

response = client.invoke_gpt_4o(input_data='what is higgs boson?',prompt="Give a one line answer.")
print(response)
```

```python
# GPT-4o, Image Processing

response = client.invoke_gpt_4o(path_to_file='<PATH_TO_FILE>', prompt="Give me two points answer.")
print(response)
```

<h3> Sample E2E code</h3>

```python
from flask import Flask, request, jsonify  
import base64  

app = Flask(__name__) 

@app.route('/upload', methods=['POST'])  
def upload_file():  
    img_file = request.files['image']  
    pdf_file = request.files['pdf']
    audio_file = request.files['audio']

    if img_file or pdf_file or audio_file:  
        response = client.invoke_gpt_4o(input_data=img_file,prompt="Give me 1 line about Carl Smith.")
        print("gpt - 4o image buffer test > \n")
        print(response)

        print("\n")
        response = client.invoke_whisper(audio_file=audio_file)
        print("whisper buffer test > ")
        print(response)

        print("\n")
        response = client.invoke_gpt_4o(input_data=pdf_file,prompt="Give me 5 points about Carl Smith.")
        print("gpt - 4o pdf buffer test > ")
        print(response)

    return "All Tests Ok"


if __name__ == '__main__':
    port = 8080
    app.run(host = "0.0.0.0", port = port)
```
